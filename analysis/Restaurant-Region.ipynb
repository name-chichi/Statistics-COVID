## 코로나 기간 동안 지역별 개업 수

```python
normal_restaurants['인허가일자'].min(), normal_restaurants['인허가일자'].max()
```

```python
normal_restaurants_5year = normal_restaurants[(normal_restaurants['인허가일자'] >= 20160101) & (normal_restaurants['인허가일자'] <= 20210630)]
```

```python
normal_restaurants_5year_opening = normal_restaurants_5year[normal_restaurants_5year['상세영업상태명'] =='영업']
```

```python
normal_restaurants_5year_opening
```

```python
normal_restaurants_5year_opening['인허가일자'] = pd.to_datetime(normal_restaurants_5year_opening['인허가일자'], format='%Y%m%d')
```

```python
normal_restaurants_5year_opening['인허가일자']
```

```python
normal_restaurants_5year_opening['year'] = normal_restaurants_5year_opening['인허가일자'].dt.year
```

```python
normal_restaurants_5year_opening['year']
```

```python
normal_restaurants_5year_opening['region']
```

```python
normal_restaurants_5year_opening['region'].unique()
```

```python
normal_restaurants_5year_opening = normal_restaurants_5year_opening[(normal_restaurants_5year_opening['region'].str.endswith('시')) | (normal_restaurants_5year_opening['region'].str.endswith('도'))]
```

```python
normal_restaurants_5year_opening['region'].unique()
```

```python
normal_restaurants_opening_region = normal_restaurants_5year_opening.groupby(['region', 'year'])
```

```python
normal_restaurants_opening_region.count()
```

```python
normal_restaurants_opening_region["상세영업상태명"].count()
```

```python
normal_restaurants_opening_region_group = normal_restaurants_opening_region["상세영업상태명"].count()
```

### COVID19 기간(19.10. ~ 21.07. ) 동안 지역별 개업 수
