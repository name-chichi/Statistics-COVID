```python
import numpy as np
```

```python
food_pub['address'] = np.where(pd.notnull(food_pub['소재지전체주소']), food_pub['소재지전체주소'], food_pub['도로명전체주소'])
```

```python
food_pub['address'] = food_pub['address'].astype(str)
```

```python
food_pub['address'] = food_pub.address.str.split(' ').str[0]
```

```python
food_pub['address']
```

## 코로나 기간 동안 지역별 개업 수

```python
food_pub_opening = food_pub[food_pub['인허가일자'] >= 20191001]
food_pub_opening = food_pub_opening[food_pub_opening['상세영업상태명'] =='영업']
```

```python
food_pub_closing = food_pub[food_pub['폐업일자'] >= 20191001]
food_pub_closing = food_pub_closing[food_pub_closing['상세영업상태명'] =='폐업']
```

```python
food_pub_opening
```

```python
food_pub_opening_address = food_pub_opening.groupby('address')
```

```python
food_pub_opening_address
```

```python
food_pub_opening_address.groups
```
