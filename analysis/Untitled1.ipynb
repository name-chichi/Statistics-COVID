```python
preprocessing_df = df
```

```python
preprocessing_df['주소'] = np.where(pd.notnull(preprocessing_df['소재지전체주소']), preprocessing_df['소재지전체주소'], preprocessing_df['도로명전체주소'])
```

```python
preprocessing_df['주소'] = preprocessing_df['주소'].astype(str)
```

```python
preprocessing_df['주소'] = preprocessing_df.주소.str.split(' ').str[0]
```

```python
preprocessing_df['주소'].unique()
```

```python
preprocessing_df = preprocessing_df[['인허가일자', '영업상태명', '폐업일자', '주소']]
```

```python
preprocessing_df
```

```python
preprocessing_df = preprocessing_df.reset_index()
```

```python
preprocessing_df.info()
```

```python
preprocessing_df = preprocessing_df.drop(['index'], axis='columns')
```

```python
preprocessing_df
```

```python
preprocessing_df['인허가일자'] = pd.to_datetime(preprocessing_df['인허가일자'], format='%Y%m%d', errors='coerce')
```

```python
preprocessing_df
```

```python
preprocessing_df['폐업일자'] = pd.to_datetime(preprocessing_df['폐업일자'], format='%Y%m%d', errors='coerce')
```
